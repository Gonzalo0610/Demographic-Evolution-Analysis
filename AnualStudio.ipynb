```python
def tabla(name):
    #Years
    ccaa = pd.DataFrame({'Periodo': range(2002, 2022)})

    #Natality Rate per 1000 people
    nat=natality[natality['Comunidades y Ciudades Autónomas']==name]
    nat=nat[nat['Nacionalidad']=='Ambas nacionalidades']
    nat['Total'] = nat['Total'].str.replace(',', '.').astype(float)
    ccaa = ccaa.merge(nat.loc[:, ['Periodo', 'Total']], on='Periodo', how='left')
    ccaa = ccaa.rename(columns={'Total': 'Natality'})

    #Mean Age of Women
    women=Edad[Edad['Comunidades y Ciudades Autónomas']==name]
    women=women[women['Sexo']=='Mujeres']
    ccaa = ccaa.merge(women.loc[:, ['Periodo', 'Total']], on='Periodo', how='left')
    ccaa = ccaa.rename(columns={'Total': 'Women_Age'})

    #Masculinity Ratio
    man=masculinity[masculinity['Comunidades y Ciudades Autónomas']==name]
    ccaa = ccaa.merge(man.loc[:, ['Periodo', 'Total']], on='Periodo', how='left')
    ccaa = ccaa.rename(columns={'Total': 'Masculinity_Ratio'})

    #Number of total Marriages
    happy=married[married['Comunidades y Ciudades Autónomas']==name]
    happy = happy.groupby(['Periodo'])['Total'].sum().reset_index()
    ccaa = ccaa.merge(happy.loc[:, ['Periodo', 'Total']], on='Periodo', how='left')
    ccaa = ccaa.rename(columns={'Total': 'Marriages'})

    # Unemployment Rate of Fertile Women
    unemployed=paradas[paradas['Comunidades y Ciudades Autónomas']==name]
    ccaa = ccaa.merge(unemployed.loc[:, ['Periodo', 'Total']], on='Periodo', how='left')
    ccaa = ccaa.rename(columns={'Total': 'Unemployment_Women'})

    #Boolean: Economic Crisis of 2008
    ccaa['Economic_Crisis'] = ccaa['Periodo'].apply(lambda x: 1 if (x >= 2009) and (x <= 2015) else 0)

    #Boolean: Arrival of trending social network
    ccaa['Instagram_Tinder'] = ccaa['Periodo'].apply(lambda x: 1 if x > 2012 else 0)

    return ccaa
```

```python
andalucia=tabla('01 Andalucía')
aragon=tabla('02 Aragón')
asturias=tabla('03 Asturias, Principado de')
baleares=tabla('04 Balears, Illes')
cantabria=tabla('06 Cantabria')
leon=tabla('07 Castilla y León')
mancha=tabla('08 Castilla - La Mancha')
catalunya=tabla('09 Cataluña')
valencia=tabla('10 Comunitat Valenciana')
extremadura=tabla('11 Extremadura')
galicia=tabla('12 Galicia')
madrid=tabla('13 Madrid, Comunidad de')
murcia=tabla('14 Murcia, Región de')
navarra=tabla('15 Navarra, Comunidad Foral de')
euskadi=tabla('16 País Vasco')
rioja=tabla('17 Rioja, La')
```

```python
def ols (var):
    return smf.ols("Natality ~  Women_Age + Masculinity_Ratio + Marriages + Unemployment_Women + Economic_Crisis + Instagram_Tinder", data=var).fit()
```
